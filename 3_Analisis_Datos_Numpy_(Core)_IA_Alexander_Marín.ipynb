```python
#Claude
import numpy as np

def generar_datos_ventas(productos, tiendas, dias):
    """
    Genera datos de ventas aleatorias para el número de productos, tiendas y días especificados
    productos (int): Número de productos
    tiendas (int): Número de tiendas
    dias (int): Número de días
    --> numpy.ndarray: Array 3D con datos de ventas aleatorias, sería el array como "datos"
    """
    return np.random.randint(0, 101, (productos, tiendas, dias))


def calcular_totales_ventas_por_producto(datos):
    """
    Calcula el total de ventas por producto a lo largo de la semana
    --> numpy.ndarray: Total de ventas por producto
    """
    return np.sum(datos, axis=(1, 2))


def calcular_totales_ventas_por_tienda(datos):
    """
    Calcula el total de ventas por tienda a lo largo de la semana
    --> numpy.ndarray: Total de ventas por tienda
    """
    return np.sum(datos, axis=(0, 2))


def calcular_promedio_ventas_por_producto(datos):
    """
    Calcula el promedio de ventas por producto por día
    --> numpy.ndarray: Promedio de ventas por producto
    """
    return np.mean(datos, axis=(1, 2))


def calcular_promedio_ventas_por_tienda(datos):
    """
    Calcula el promedio de ventas por tienda por día
    --> numpy.ndarray: Promedio de ventas por tienda
    """
    return np.mean(datos, axis=(0, 2))


def encontrar_producto_mayor_menor_ventas(totales_por_producto):
    """
    Encuentra el producto con mayor y menor ventas totales en la semana
    --> tuple: Índices del producto con mayor y menor ventas
    """
    producto_mayor_ventas = np.argmax(totales_por_producto)
    producto_menor_ventas = np.argmin(totales_por_producto)
    return producto_mayor_ventas, producto_menor_ventas


def encontrar_tienda_mayor_menor_ventas(totales_por_tienda):
    """
    Encuentra la tienda con mayor y menor ventas totales en la semana
    --> tuple: Índices de la tienda con mayor y menor ventas
    """
    tienda_mayor_ventas = np.argmax(totales_por_tienda)
    tienda_menor_ventas = np.argmin(totales_por_tienda)
    return tienda_mayor_ventas, tienda_menor_ventas


def main():
    """
    Función principal que ejecuta todo el análisis de ventas
    """
    # Configuración de parámetros
    productos = 10
    tiendas = 5
    dias = 7

    # Genera los datos de ventas
    datos = generar_datos_ventas(productos, tiendas, dias)

    # Calcula los totales y promedios
    totales_por_producto = calcular_totales_ventas_por_producto(datos)
    totales_por_tienda = calcular_totales_ventas_por_tienda(datos)
    promedio_por_producto = calcular_promedio_ventas_por_producto(datos)
    promedio_por_tienda = calcular_promedio_ventas_por_tienda(datos)

    # Encuentra el producto y la tienda con mayor y menor ventas
    producto_mayor_ventas, producto_menor_ventas = encontrar_producto_mayor_menor_ventas(totales_por_producto)
    tienda_mayor_ventas, tienda_menor_ventas = encontrar_tienda_mayor_menor_ventas(totales_por_tienda)

    # Imprime los resultados
    print("Total de ventas por producto a lo largo de la semana:", totales_por_producto)
    print("Total de ventas por tienda a lo largo de la semana:", totales_por_tienda)
    print("Promedio de ventas por producto por día:", promedio_por_producto)
    print("Promedio de ventas por tienda por día:", promedio_por_tienda)
    print(f"Producto con mayor ventas: Producto {producto_mayor_ventas}")
    print(f"Producto con menor ventas: Producto {producto_menor_ventas}")
    print(f"Tienda con mayor ventas: Tienda {tienda_mayor_ventas}")
    print(f"Tienda con menor ventas: Tienda {tienda_menor_ventas}")

if __name__ == "__main__":
    main()
```

```python
#Claude
import numpy as np
import matplotlib.pyplot as plt

def generar_datos_ventas(productos, tiendas, dias):
    """Genera datos de ventas aleatorias"""
    return np.random.randint(0, 101, (productos, tiendas, dias))

def visualizar_ventas_por_producto(datos):
    """
    Crea un gráfico de barras para mostrar ventas totales por producto

    Args:
    datos (numpy.ndarray): Array 3D de datos de ventas
    """
    # Calcular ventas totales por producto
    ventas_por_producto = np.sum(datos, axis=(1, 2))

    plt.figure(figsize=(12, 6))
    plt.bar(range(len(ventas_por_producto)), ventas_por_producto)
    plt.title('Ventas Totales por Producto')
    plt.xlabel('Producto')
    plt.ylabel('Ventas Totales')
    plt.xticks(range(len(ventas_por_producto)), [f'Producto {i}' for i in range(len(ventas_por_producto))])

    # Añadir valores en la parte superior de cada barra
    for i, v in enumerate(ventas_por_producto):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

def visualizar_ventas_diarias_por_tienda(datos):
    """
    Crea un gráfico de líneas para mostrar ventas diarias por tienda

    Args:
    datos (numpy.ndarray): Array 3D de datos de ventas
    """
    plt.figure(figsize=(12, 6))

    # Calcular ventas por tienda para cada día
    ventas_diarias = np.sum(datos, axis=0)

    for tienda in range(ventas_diarias.shape[0]):
        plt.plot(range(1, ventas_diarias.shape[1] + 1),
                 ventas_diarias[tienda],
                 marker='o',
                 label=f'Tienda {tienda}')

    plt.title('Ventas Diarias por Tienda')
    plt.xlabel('Día')
    plt.ylabel('Ventas')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

def visualizar_ventas_por_tienda(datos):
    """
    Crea un gráfico de barras para mostrar ventas totales por tienda

    Args:
    datos (numpy.ndarray): Array 3D de datos de ventas
    """
    # Calcular ventas totales por tienda
    ventas_por_tienda = np.sum(datos, axis=(0, 2))

    plt.figure(figsize=(12, 6))
    plt.bar(range(len(ventas_por_tienda)), ventas_por_tienda)
    plt.title('Ventas Totales por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Ventas Totales')
    plt.xticks(range(len(ventas_por_tienda)), [f'Tienda {i}' for i in range(len(ventas_por_tienda))])

    # Añadir valores en la parte superior de cada barra
    for i, v in enumerate(ventas_por_tienda):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

def crear_tabla_ventas(datos):
    """
    Crea una tabla de ventas con pandas

    Args:
    datos (numpy.ndarray): Array 3D de datos de ventas
    """
    import pandas as pd

    # Calcular ventas totales por producto y tienda para cada día
    tabla_ventas = []
    for dia in range(datos.shape[2]):
        ventas_dia = datos[:, :, dia]
        tabla_dia = pd.DataFrame(ventas_dia,
                                 columns=[f'Tienda {i}' for i in range(datos.shape[1])],
                                 index=[f'Producto {i}' for i in range(datos.shape[0])])
        tabla_dia['Total Día'] = tabla_dia.sum(axis=1)
        tabla_ventas.append(tabla_dia)

    # Imprimir tablas de cada día
    for dia, tabla in enumerate(tabla_ventas, 1):
        print(f"\nVentas del Día {dia}:")
        print(tabla)
        print("\n" + "="*50)

def main():
    # Configuración de parámetros
    productos = 10
    tiendas = 5
    dias = 7

    # Genera los datos de ventas
    datos = generar_datos_ventas(productos, tiendas, dias)

    # Visualizaciones
    visualizar_ventas_por_producto(datos)
    visualizar_ventas_por_tienda(datos)
    visualizar_ventas_diarias_por_tienda(datos)

    # Tabla de ventas
    crear_tabla_ventas(datos)

if __name__ == "__main__":
    main()
```
