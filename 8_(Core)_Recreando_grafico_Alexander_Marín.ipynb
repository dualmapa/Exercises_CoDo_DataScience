```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('tips')
```

#2) **Creación de la Figura y el Eje:**
Utiliza plt.subplots para crear una figura de tamaño adecuado.



```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Mostrar el gráfico
plt.show()
```

#3) **Gráfico de Dispersión:**
Completa la llamada a sns.scatterplot añadiendo las columnas correctas para los parámetros hue, style y s, y asegúrate de asignar el gráfico al eje correcto (ax).



```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(13, 6))
# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')

# Mostrar el gráfico
plt.show()
```

#4) **Cuadrícula:**
Asegúrate de que la cuadrícula esté configurada correctamente utilizando los parámetros which, linestyle, linewidth y color.



```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(13, 6))
# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper left', fontsize='medium', frameon=False, shadow=False, title='Detalles')
# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
# Cambiar color al fondo del gráfico
ax.set_facecolor("lightgray")

# Mostrar el gráfico
plt.show()
```

#5) **Anotaciones:**
Completa las coordenadas xy y xytext para ubicar la anotación correctamente.



```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(13, 6))
# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax,
                palette=('darkred','royalblue'))
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper left', fontsize='medium', frameon=False, shadow=False, title='Detalles')
# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
# Cambiar color al fondo del gráfico
ax.set_facecolor("lightgray")

# Añadir anotación
ax.annotate('Mayor tip en la cena', xy=(51, 10), xytext=(40, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Mostrar el gráfico
plt.show()
```

#6) **Flechas:**

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(13, 6))
# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax,
                palette=('darkred','royalblue'))
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper left', fontsize='medium', frameon=False, shadow=False, title='Detalles')
# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
# Cambiar color al fondo del gráfico
ax.set_facecolor("lightgray")
# Añadir anotación
ax.annotate('Mayor tip en la cena', xy=(51, 10), xytext=(40, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha
ax.arrow(30, 5, 10, 3, head_width=0.8, head_length=0.3, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
